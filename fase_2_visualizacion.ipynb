```python
#importamos librerias
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
```

```python
#leemos .csv limpio:
df_custom_data = pd.read_csv("customer_dataset.csv")
df_custom_data.head()
```

```python
df_custom_data.info()
```

1. ¿Cómo se distribuye la cantidad de vuelos reservados por mes durante el año?

```python
#columnas que usaremos: 'Month', 'Year' y 'Flights Booked'.
#usamos Lineplot porque queremos visualizar la evolucion en el timepo.
#de esta forma vemos visualmente los dos años a lo largo de lso meses.
#creamos lista con los meses del año para definir el orden correcto en la visualización.
#le damos rotación para que no se pisen unas a otras.
meses = ['January', 'February', 'March', 'April', 'May', 'June', 
               'July', 'August', 'September', 'October', 'November', 'December']

sns.lineplot(   x = "Month",
                y = "Flights Booked",
                hue = "Year",
                data = df_custom_data,
                palette = ["blue", "yellow"])

plt.title("Flights booked per month")

plt.xticks(ticks = range(len(meses)), labels = meses, rotation=90)
```

Resultado visualización:

- 2017 y 2018 siguen una tendencia similar.

- Picos de reserva: febrero, agosto, diciembre.

- Coinciden en caídas de reservas (marzo, junio y noviembre).

- En 2018 hay un número de reservas más elevado en comparación con 2017.

2. ¿Existe una relación entre la distancia de los vuelos y los puntos acumulados por
los clientes?

```python
#columnas que usaremos: 'Distance' y 'Points Accumulated'.
#usamos scatterplot para ver la relación entre dos variables numéricas.
#con scatterplot vemos si hay correlación directa.

sns.scatterplot(    x='Distance',
                    y='Points Accumulated',
                    data = df_custom_data,
                    color = 'green'
                    )

plt.title("Relationship between Flight distance and points accumulated")
```

Resultado visualización:

- Hay una relación lineal positiva entre distancia de vuelos y puntos acumulados.

- Cuanto mayor es la distancia de los vuelos, mayor es el número de puntos acumulados.

- En conclusión, los puntos estan directamente relacionados con la distancia de los vuelos.

3. ¿Cuál es la distribución de los clientes por provincia o estado?

```python
#columna que usaremos: 'Province'.
#usamos countplot para ver de forma sencilla las ocurrencias por provincia.
sns.countplot(    x='Province',
                    data = df_custom_data,
                    color = 'red'
                    )

plt.title("Customer distribution by province")
plt.xticks(rotation=90)
```

Resultado visualización:

- Visualizamos cada barra por provincia para ver la distribución de la población.
- Las barras más altas (Ontario, British Columbia y Quebec) representan el mayor número de clientes, de esta forma podemos identificar la distribución geográfica con más claridad.

4. ¿Cómo se compara el salario promedio entre los diferentes niveles educativos de
los clientes?

```python
#columnas que usaremos: 'Salary' y 'Education'.
#usamos barplot para visualizar el promedio de cada nivel educativo.
#'estimator=np.mean' calcula el salario promedio para cada nivel educativo.
#vamos a crear una lista para tener ordenados los niveles educativos:
niveles_educativos = ['High School or Below', 'College', 'Bachelor', 'Master', 'Doctor']

#convertimos la columna 'Education' en una categoría ordenada
df_custom_data['Education'] = pd.Categorical(df_custom_data['Education'], categories=niveles_educativos, ordered=True)


sns.barplot(   x = "Education",
                y = "Salary",
                data = df_custom_data,
                estimator=np.mean)

plt.title("Comparison of average salary by education level")
plt.xticks(rotation=90)
```

Resultado visualización:

- Los clientes con un 'Doctor' tienen el salario promedio más alto.
- Los clientes con un nivel educativo 'Bachelor' o inferior tienen los salarios promedio bajo.
- Visualizamos un aumento del salario promedio conforme va aumentando el nivel educativo.

5. ¿Cuál es la proporción de clientes con diferentes tipos de tarjetas de fidelidad?

```python
#columna que vamos a usar: 'Loyalty Card'.
#countplot es perfecto para visualizar con claridad.
sns.countplot(  x='Loyalty Card',
                data=df_custom_data,
                palette = ["blue", "yellow", "green"])


plt.title('Proportion of customers by loyalty card type')
plt.xticks(rotation=90)
```

Resultado visualización:

- Tarjeta Star la más popular.
- Clientes con tarjeta Star: probablemente más leales y frecuentes, debido a su relación con la acumulación de puntos.
- Las tarjetas Nova y Aurora tienen menos usuarios, lo que puede reflejar diferencias en los beneficios ofrecidos o en el perfil de los clientes.


6. ¿Cómo se distribuyen los clientes según su estado civil y género?

```python
#usamos columnas 'Gender' y 'Marital Status'.
#usamos countplot con hue para mostrar separación por género.
sns.countplot(  x = "Marital Status",
                data = df_custom_data,
                hue = "Gender"
                )

plt.title("Number of clients by marital status")
plt.xticks(rotation = 90)
```

Resultado visualización:

- Proporción equilibrada entre hombres y mujeres en la distribución por estado civil.
- Se peude visualizar una pequeña diferencia en los clientes solteros y divorciados, siendo las mujeres ligeramente más representadas en el grupo de solteros.
